```python
import numpy as np
from matplotlib import pyplot as plt
```

Generate some random data

```python
N = 100

# randn will create N random numbers under 'normal' distribution
x = np.random.randn(N)
# use x to create y and add some more randomness
y = x*3 + np.random.randn(N)*2
```

Plot what we have

```python
plt.scatter(x,y)
plt.show()
```

Let's see how the data are distributed using a **histogram**

```python
plt.hist(x, bins=20, alpha=0.5) # alpha=0.5 makes the plot transparent (50%) so that we can see both
plt.hist(y, bins=20, alpha=0.5)
plt.show()
```

What if you wanted to visualize two series? For example, this could be two fields in your survey, side by side.

We'll generate `y1` and `y2` by taking $\log(x)$ with different levels of random noise: `np.random.randn(N)*0.75`

```python
N = 1000

x = np.random.randn(N)
y1 = np.log(x) + np.random.randn(N)*0.75
y2 = np.log(x) + np.random.randn(N)*2.2

plt.scatter(x,y1, alpha=0.5)
plt.scatter(x,y2, alpha=0.5)
plt.show()
```

Or you could plot `y1` and `y2` like so... (this is the same thing as the plot above, just side by side now)

```python
plt.figure(figsize=(12,5))

# first subplot (on the left)
plt.subplot(1,2,1)
plt.scatter(x,y1, alpha=0.5, label='y1 = log(x) + noise')
plt.xlabel('x')
plt.ylabel('y1')
plt.legend()

# second subplot - on the right
plt.subplot(1,2,2)
plt.scatter(x,y2, alpha=0.5, label='y1 = log(x) + more noise')
plt.xlabel('x')
plt.ylabel('y2')
plt.legend()

plt.show()
```

So far, we've only plotted *continuous* data, or *metrics*...

In your survey data, you will have a number of categorical variables, or *attributes*. Let's generate some data and plot them...

```python
N = 100

categories = ['Freshman', 'Sophomore', 'Junior', 'Senior']

# randomly choose between our categories with a total size of N (100)
a = np.random.choice(categories, N)
# np.unique with the return_counts=True option gives us the categories with the count of each in our data, `a`
a_categories, a_counts = np.unique(a, return_counts=True)
```

Let's plot our data using `plt.bar()`

```python
plt.bar(a_categories, a_counts)
plt.show()
```
